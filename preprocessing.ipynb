```python
import pandas as pd
df = pd.read_csv('Reviews_Collected.csv')
```

```python
df
```

```python
# Removing reviews other than english
df = df.drop([0,1,2,3,4,8,10,11,12,17,18,19,20,21,22,26,28,29,31,32])
```

```python
df = df.reset_index(drop=True)
df
```

```python
df = df.rename(columns={'0': 'Reviews'})
```

```python
df.head()
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
words_descriptions = df['Reviews'].apply(tokenizer.tokenize)
words_descriptions.head()
```

```python
all_words = [word for tokens in words_descriptions for word in tokens]
df['description_lengths']= [len(tokens) for tokens in words_descriptions]
VOCAB = sorted(list(set(all_words)))
print("%s words total, with a vocabulary size of %s" % (len(all_words), len(VOCAB)))
```

```python
df.head()
```

```python
# Checking most common words
from collections import Counter
count_all_words = Counter(all_words)
count_all_words.most_common(100)
```
